# Hypotenuses

# $$hypotenuse = \sqrt{leg1^2 + leg2^2}$$
# $$hypotenuse = \sqrt{2 * leg^2}$$

## Import modules

```python
import math
import numpy as np
```

## Calculate Hypotenuses without NumPy

```python
legs = range(1, 101)
hypotenuses = [(math.sqrt(2 * leg**2)) for leg in legs]
hypotenuses
```

## Calculate Hypotenuses with NumPy

```python
r = np.arange(1, 101)
hypotenuses = np.hypot(r, r)
hypotenuses
```

## Compare Efficiency

```python
%%timeit -n 1000
legs = range(1, 101)
hypotenuses = [(math.sqrt(2 * leg**2)) for leg in legs]
```

```python
%%timeit -n 1000
r = np.arange(1, 101)
hypotenuses = np.hypot(r, r)
```
